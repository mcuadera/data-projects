### Core functions
Throughout the project, probabilities and combinations will be calculated often. The initial step is to write two functions:
- A function that calculates factorials; and
- A function that calculates combinations.

```python
def factorial(n):
    final = 1
    for i in range(1, n+1):
        final *= i
    return final

def combinations(n, k):
    return factorial(n)/(factorial(k)*factorial(n-k))
```

### One-ticket probability
On this section, the focus is on writing a function that calculates the probability of winning the big prize.

In the 6/49 lottery, six numbers are drawn from a set of 49 numbers that range from 1 to 49. A player wins the big prize if the six numbers on their tickets match all the six numbers drawn. If a player has a ticket with the numbers {13, 22, 24, 27, 42, 44}, he only wins the big prize if the numbers drawn are {13, 22, 24, 27, 42, 44}. If only one number differs, the player doesn't win.

Players should be able to calculate the probability of winning the big prize with the various numbers they play on a single ticket (for each ticket a player chooses six numbers out of 49). Thus, a function should be built that calculates the probability of winning the big prize for any given ticket.

In the app, these should also be considered:

- Inside the app, the user inputs six different numbers from 1 to 49.
- Under the hood, the six numbers will come as a Python list, which will serve as the single input to our function.
- The function should print the probability value in a friendly way â€” in a way that people without any probability training are able to understand.

```python
def one_ticket_probability(a_list):
    total_outcomes = combinations(49, len(a_list))
    probability = 1/total_outcomes
    return print('The probability of winning with {} is {:.6%}'.format(a_list, probability))
```

```python
one_ticket_probability([13, 22, 24, 27, 42, 44])
```

### Multi-Ticket Probability
Lottery addicts usually play more than one ticket on a single drawing, thinking that this might increase their chances of winning significantly. Since the purpose of the app is to help them better estimate their chances of winning, I built function that will allow the users to calculate the chances of winning for any number of different tickets.

Additional considerations:

- The user will input the number of different tickets they want to play (without inputting the specific combinations they intend to play).
- Our function will see an integer between 1 and 13,983,816 (the maximum number of different tickets). Other lottery will have different number of tickets, but that is beyond the scope of the current project.
- The function should print information about the probability of winning the big prize depending on the number of different tickets played.

```python
def multi_ticket_probability(n):
    tot_outcomes = combinations(49,6)
    probability = n/tot_outcomes
    print('The likelihood of winning with {} tickets is {:.6%}'.format(n, probability))
```

### Less Winning Numbers
In most 6/49 lotteries, there are smaller prizes if a player's ticket match two, three, four, or five of the six numbers drawn. As a consequence, the users might be interested in knowing the probability of having two, three, four, or five winning numbers.

Additional considerations:

- Inside the app, the user inputs:
    - six different numbers from 1 to 49; and
    - an integer between 2 and 5 that represents the number of winning numbers expected
- The function should print information about the probability of having the inputted number of winning numbers.

What if the user wants to calculate the probability of having exactly two winning numbers? I needed to figure out the number of combinations that would contain two numbers from a list of six numbers.

$\binom{\text{total numbers to draw}}{\text{sample}} = \binom{6}{2} = \frac{6!}{2!(6-2)!} = 15$

The last four numbers need to be incorrect. However, the winning numbers from the pool should be removed so that our pool only contains the incorrect numbers:

$\text{total numbers} - \text{winning numbers} = 49 - 6 = 43$

From the 43, I needed to calculate the total number of four number combinations.

$\binom{\text{total incorrect numbers}}{6 - \text{total numbers left to draw}} = \binom{43}{4} = 123,410$

Since there are 15 ways to get two winning number from drawing six numbers, **AND** there are 123,410 ways of getting the rest incorrect, we need to multiply them to get the total number of six number combinations containing at most two winning numbers.

```python
def probability_less_6(n=2):
    n_combinations_ticket = combinations(6, n)
    n_combinations_remaining = combinations(43, 6 - n)
    successful_outcomes = n_combinations_ticket * n_combinations_remaining
    total_outcomes = combinations(49,6)
    prob = successful_outcomes / total_outcomes
    if 2 <= n <=5: 
        print('The likelihood of having a winning {}-number combination is: {:.6%}'.format(n, prob))
    else:
        print('Pick a number between 2 and 5')
```

```python
test_inputs = [2, 3, 4, 5]

for i in test_inputs:
    probability_less_6(i)
```
